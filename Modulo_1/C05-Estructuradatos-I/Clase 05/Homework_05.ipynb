```python
class pilas:
    def __init__(self):
       self.pila = []
      
       #self.elementos = 0
    
    def push(self,element):
        self.pila.append(element)
        #self.elementos += 1
        return(element)
    
    def pop(self):
        if len(self.pila) <= 0:
            ultimo = None
        else:      
            ultimo = self.pila[-1]
            self.pila.pop(-1)
        return(ultimo)
    
    def numElement(self):     
        return(len(self.pila))
    
    def getElements(self):
        return(self.pila)

```

```python
#Juego de los 5 numeros
import random as rd
class juego1:
    def __init__(self):
        self.mazo = pilas()
        self.mano = []
        self.countNumeros = 0
        self.ganados =0
        self.perdidos =0
        self.puntuacion =0
        self.first=True
        for i in range(19):
            self.mazo.push(rd.randint(1,20))
        print(self.mazo.getElements())
        return
         
    def repartir(self):
        num = self.mazo.pop()
        print("Numeros que quedan: ",self.mazo.getElements())
        self.mano.append(num)
        self.countNumeros += 1
        print(f"numeros en mi mano: {self.miMano()}")
        return num
    
    def fPuntuacion(self):
        total = sum(self.mano)
        if total > 50:           
            self.puntuacion += total
            print("*******    Perdiste **********")
            print(f"Tu total de puntos fue {total} que se exedio en {total-50} puntos del máximo (50)")
            print()
        else:
            print("*******  Ganaste **********")
            print()
            tot = 10 - (50 - total)
            self.puntuacion += tot
            print(f"Tus puntos de la partida son {tot} que suman {total} puntos totales")
            self.fEstadisticas()
        return

        
    def fEstadisticas(self):
        print("Tus estadistcas son:")
        print("----------")    
        print("Partidas ganadas: ",self.ganados)
        print("Partidas Perdidas: ",self.perdidos)
        print("Puntuación: ", self.puntuacion)
        return
    
    def miMano(self):
        return(self.mano)

    def menuJuego(self):
        opcion =0
        while True:
            if self.first:
                print("#############  ¡  COMENZAMOS !  ################")
                self.first = False
            else:
                print("###########  ¡  Sigiente numero !  #############")

            print()
            print("Seleccione una opcion:")
            print("a). Pedir un numero")
            print("b). Ver mis numeros")
            print("c). No mas numeros y cerrar")
            print("----------------------------------------------")
            sel = input()
            if sel == 'a' or sel == 'A' or sel == '1':
                self.repartir()
                print(f"Con este numero tienes {self.countNumeros} numeros seleccionados y son:" )
                print(self.miMano())
                print(f"haciendo un total de {sum(self.mano)} puntos")
                opcion = 1
            elif sel == 'b' or sel == "B" or sel == '2':
                print(f"Con este numero tienes {self.countNumeros} numeros seleccionados y son:" )
                print(self.miMano())
                print(f"haciendo un total de {sum(self.mano)} puntos")
                opcion = 2
            elif sel == 'c' or sel == "C" or sel == '3':
                self.fPuntuacion()
                opcion = 3
                break
            else:
                print(" --------  Opcion Invalida -----------------")
                continue
        return opcion
        
    def menuInicio(self):
        print("1. Nuevo juego")
        print("2. exit")
        print("3. estadisticas")
        sel = input("Introduce tu opcion (1,2,ó 3)")
        print("*****************************************")
        if(sel== '1' ):
            return(1)
        elif sel== '2':
            return(2)
        elif sel== '3':
            self.fEstadisticas()
            return(3)
        else:
            return(-1)
              
```

2) Implementar un juego donde constas de 2 jarras, de capacidad 5 y 3 litros respectivamente, y debes colocar 4 litros en la jarra de 5L.
Las opciones posibles son:
* Llenar la jarra de 3 litros
* Llenar la jarra de 5 litros
* Vaciar la jarra de 3 litros
* Vaciar la jarra de 5 litros
* Verter el contenido de la jarra de 3 litros en la de 5 litros
* Verter el contenido de la jarra de 5 litros en la de 3 litros

```python
class lasJarras:
    def __init__(self):
         self.jar3l = 0
         self.jar5l = 0
         return
    
    def full_jar3l(self):
        sobra =0 
        self.jar3l += 3
        if self.jar3l > 3:
            sobra = self.jar3l - 3
            self.jar3l = 3
        return sobra
    
    def full_jar5l(self):
        sobra = 0
        self.jar5l += 5
        if self.jar5l > 5:
            sobra = self.jar5l -5
            self.jar5l = 5
        return sobra
    
    def empty_jar3l(self):
        sobra = self.jar3l
        self.jar3l = 0;
        return sobra
    
    def empty_jar5l(self):
        sobra = self.jar5l
        self.jar5l = 0;
        return sobra
    
    def fill_jar3l(self,lt):
        sobra = 0
        self.jar3l += lt
        if self.jar3l > 3:
            sobra = self.jar3l -3
            self.jar3l = 3
        return sobra
    
    def fill_jar5l(self,lt):
        sobra = 0
        self.jar5l += lt
        if self.jar5l > 5:
            sobra = self.jar5l -5
            self.jar5l = 5
        return sobra
    
    def get_cantidad_jar3l(self):
        return(self.jar3l)
    
    def get_cantidad_jar5l(self):
        return(self.jar5l)
    
```

```python
class Jarra:
    def __init__(self, capacidad):
        self.capacidad = capacidad
        self.cantidad = 0

    def llenar(self):
        self.cantidad = self.capacidad

    def vaciar(self):
        self.cantidad = 0

    def verter_en(self, otra_jarra):
        cantidad_total = self.cantidad + otra_jarra.cantidad
        if cantidad_total > otra_jarra.capacidad:
            self.cantidad = cantidad_total - otra_jarra.capacidad
            otra_jarra.cantidad = otra_jarra.capacidad
        else:
            self.cantidad = 0
            otra_jarra.cantidad = cantidad_total


class Juego:
    def __init__(self):
        self.jarra_5l = Jarra(5)
        self.jarra_3l = Jarra(3)

    def imprimir_estado(self):
        print(f"Jarra de 5L: {self.jarra_5l.cantidad}/{self.jarra_5l.capacidad}L")
        print(f"Jarra de 3L: {self.jarra_3l.cantidad}/{self.jarra_3l.capacidad}L")

    def llenar_jarra_5l(self):
        self.jarra_5l.llenar()

    def llenar_jarra_3l(self):
        self.jarra_3l.llenar()

    def vaciar_jarra_5l(self):
        self.jarra_5l.vaciar()

    def vaciar_jarra_3l(self):
        self.jarra_3l.vaciar()

    def verter_jarra_3l_en_5l(self):
        self.jarra_3l.verter_en(self.jarra_5l)

    def verter_jarra_5l_en_3l(self):
        self.jarra_5l.verter_en(self.jarra_3l)

    def jugar(self):
        print("Bienvenido al juego de las jarras!")
        self.imprimir_estado()
        print("Objetivo: colocar 4L en la jarra de 5L")

        while self.jarra_5l.cantidad != 4:
            print("Elige una acción:")
            print("1. Llenar jarra de 3L")
            print("2. Llenar jarra de 5L")
            print("3. Vaciar jarra de 3L")
            print("4. Vaciar jarra de 5L")
            print("5. Verter jarra de 3L en jarra de 5L")
            print("6. Verter jarra de 5L en jarra de 3L")
            accion = int(input("Acción: "))

            if accion == 1:
                self.llenar_jarra_3l()
            elif accion == 2:
                self.llenar_jarra_5l()
            elif accion == 3:
                self.vaciar_jarra_3l()
            elif accion == 4:
                self.vaciar_jarra_5l()
            elif accion == 5:
                self.verter_jarra_3l_en_5l()
            elif accion == 6:
                self.verter_jarra_5l_en_3l()

            self.imprimir_estado()

        print("¡Felicidades! Has colocado 4L en la jarra de 5L.")

juego = Juego()
juego.jugar()
```
